```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame(np.full((2, 3), "x", dtype="object"), columns=["A", "B", "C"])
df1
```

```python
df2 = pd.DataFrame(np.full((3, 3), "o", dtype="object"), columns=["A", "B", "C"])
df2
```

```python
df3 = pd.DataFrame(np.full((2, 2), "v", dtype="object"), columns=["D", "E"])
df3
```

```python
pd.concat([df1, df2])
```

```python
# reset index as well

pd.concat([df1, df2]).reset_index(drop=True)
```

```python
# pandas fills in values for df that doesnt match nan

pd.concat([df1, df3])
```

```python
# using keys to maintain original dfs

df4 = pd.concat([df1, df2], keys=["df1", "df2"])
df4
```

```python
# now can recall previous rows using this

df4.loc["df2", :]
```

```python
# concantenate along axis 1

pd.concat([df1, df3], axis=1)
```

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df2], axis=1, join="inner")
```
