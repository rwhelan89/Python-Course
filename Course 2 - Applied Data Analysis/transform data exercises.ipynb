```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("c2_cereal.csv")

print('Data shape: ',df.shape)
```

```python
df.head()
```

```python
df["manufacturer"] = df.mfr.map({"A": "American Home Food Products", "G": "General Mills", "K": "Kellogs", "N": "Nabisco", "P": "Post", "Q": "Quaker Oats", "R": "Ralston Purina"})
df.head()
```

```python
def conversion(x): 
    y = x * 4.184
    return y

df["kilojoules"] = df.calories.apply(conversion)
df.head()
```

```python
df_num = df.loc[:, ["calories", "protein", "fat", "sodium", "fiber", "carbo", "sugars", "potass", "vitamins", "shelf", "weight", "cups", "rating", "kilojoules"]]
df_num.head()
```
