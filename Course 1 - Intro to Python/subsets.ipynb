```python
df = pd.DataFrame(
    {
        "user": [1, 2, 3],
        "age": [24, 54, 17],
        "sex": ["F", "F", "M"],
        "occupation": ["technician", "musician", "student"],
    }
)
df
```

```python
df[df.sex == "F"].sex
```

```python
df[df.sex == "F"].sex = "Female"
```

```python
#hasnt changed the original data
df
```

```python
#to avoid chained assignments use .loc instead
df.loc[df.sex == "F", "sex"] = "Female"
df
```

```python
#when defining a new copy ensure to use the function
df2 = df.loc[df.sex == "F"].copy()
```

```python
df2.loc[0:1, "sex"] = "Female"
df2
```

```python
df = pd.DataFrame(
    {
        "user": [1, 2, 3],
        "age": [24, 54, 17],
        "sex": ["F", "F", "M"],
        "occupation": ["technician", "musician", "student"],
    }
)
df
```

```python
#map function for efficient. have to make sure specify for all variables ie cant just replace F have to do M too
df["sex"] = df["sex"].map({"F": "Female", "M": "Male"})
df
```

```python
#replace function
df["sex"].replace("Female", "1")
```

```python
#applymap to apply a function
def my_func(x):
    if x > 5:
        size = "Large"
    elif x > 3:
        size = "Medium"
    else:
        size = "Small"
    return size
```

```python
df2.applymap(my_func)
```
