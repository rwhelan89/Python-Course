```python
#isin function conditional based on values in list specified
class_23 = titanic[titanic["pclass"].isin([2,3])]
```

```python
class_23.head()
```

```python
#removing unknown age results. notna() conditional function returns true boolean for each row that isnt null
age_no_na = titanic[titanic["age"].notna()]
```

```python
age_no_na.head()
```

```python
age_no_na.shape
```

```python
#using loc and iloc function which allows to pull rows and column in one function
over35 = titanic.loc[titanic["age"] > 35, "sex"]
```

```python
over35.head()
```

```python
over35.shape
```

```python
#iloc you can specify specific rows/columns
titanic.iloc[9:25, 2:5]
```

```python
#allows to specify values to multiple entries
titanic.iloc[0:3, 3] = "anonymous"
```
