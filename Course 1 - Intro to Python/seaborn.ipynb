```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
sns.histplot(iris["sepal_length"], kde=True, stat="density")
plt.show()
```

```python
sns.regplot(x=iris.sepal_length, y=iris.petal_length)
plt.title("A Title")
sns.set_style("whitegrid")
plt.show()
```

```python
sns.pairplot(iris, hue="species", corner=True, height=1.5)
plt.show()
```

```python
sns.pairplot(iris, x_vars=["sepal_length", "sepal_width"],
                   y_vars=["petal_length"],
                   hue="species",
                   height=3,
                   palette="pastel")
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
grid = sns.FacetGrid(tips, col="time") #define grid and conditioning variable
grid = grid.map(sns.scatterplot, "total_bill", "tip")
```

```python
grid = sns.FacetGrid(tips, col="day") #define grid and conditioning variable
grid = grid.map(sns.scatterplot, "total_bill", "tip")
```

```python
grid = sns.FacetGrid(tips, col="day", hue="sex") #add on colour code variable
grid = grid.map(sns.scatterplot, "total_bill", "tip")
grid.add_legend()
plt.show()
```

```python
grid = sns.FacetGrid(tips, col="time", row="sex") #define 1st and 2nd variable
grid = grid.map(sns.scatterplot, "total_bill", "tip")
grid.add_legend()
plt.show()
```

```python
grid = sns.FacetGrid(tips, col="time", row="sex", hue="smoker") #define 1st and 2nd variable
grid = grid.map(sns.scatterplot, "total_bill", "tip")
grid.add_legend()
plt.show()
```

```python
sns.violinplot(x=iris["species"], y="petal_length", data=iris)
plt.title("A violin plot of petal length")
plt.show()
```
