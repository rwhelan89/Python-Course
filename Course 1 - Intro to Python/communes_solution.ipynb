Section B

1. Display mean, minimum and max

```python
# use describe function however only display the 3 required values

df.describe().loc[["mean", "min", "max"]]    
```

2. List top 10 most populated communes

```python
# select index commune 

df = df.reset_index().set_index("Commune")     

# sort by residents in ascending order

df = df.sort_values(by="Residents", ascending=False)   

# display top 10

df["Residents"].head(10)                               
```

3. List bottom 10 populated communes

```python
# sort by residents & display top 10

df = df.sort_values(by="Residents")    
df["Residents"].head(10)                
```

4. Group communes by Canton and save as separate .csv files

```python
# group cantons within df and assign to individual csv files

for name, group in df.groupby('Canton'):   
    group.to_csv('{}.csv'.format(name))   
```

5. Compute mean probability density for each commune listing ascendingly

```python
# use grouping to calculate mean population densities of cantons & list

mean_density = df.groupby('Canton')["Population density per km²"].mean()  
mean_density.sort_values(ascending=False)                                 
```

6. Number of communes within each Canton > 50% pop between 20 - 64

```python
# filter for all communes with higher than 50% in age range

middle_age_majority = df[df['20-64 years'] > 50]           

# group by canton and count number of commmunes

middle_age_majority_sum = middle_age_majority.groupby('Canton').count()       

# display list

middle_age_majority_sum['20-64 years']                                        
```

7. Difference between max and min elevations for each canton

```python
#group by cantons and find min/max values

elevation_max = df.groupby('Canton').max()                      
elevation_min = df.groupby('Canton').min()         

# subtract differences

elevation_range = elevation_max['Elevation'] - elevation_min['Elevation']     

# rank top 5 in ascending order

elevation_range.sort_values(ascending=False).head(5)            
```

C. Exploration: Visualisations

```python
# import matplot for this section

import matplotlib.pyplot as plt      
%matplotlib inline
```

1. Top 10 populated communes in bar chart

```python
# extract residents data and sort values from high to low
# create dataset of only top 10 values

population = df['Residents'].sort_values(ascending=False)
top_population = population.head(10)

# build horizontal bar chart with selected data

colormap = plt.cm.tab10(range(0,len(top_population)))
bar_chart = top_population.plot.barh(color=colormap, edgecolor='black')

# annotate chart

plt.xlabel('Residents')
plt.title('Top 10 most populated Communes in Switzerland')
plt.xticks(fontsize=9)
plt.show(bar_chart)
```

2. Horizontal bar chart of land division of top 10 most populated communes

```python
# extract dataset with selected columns
# create subset of top 10
# drop residents column as not required for chart

pop_land_type = df[['Residents', 'Settlement area', 'Agricultural area', 'Wooded area', 'Unproductive area']]
pop_land_type_top = pop_land_type.sort_values('Residents', ascending=False).head(10)
land_type_top = pop_land_type_top.drop(['Residents'], axis=1)

# start building blocks for the stacked chart

settlement_area = land_type_top['Settlement area']
agricultural_area = land_type_top['Agricultural area']
wooded_area = land_type_top['Wooded area']
unproductive_area = land_type_top['Unproductive area']

# create horizontal barchart stacking from each previous land type

settlement_area.plot.barh()
agricultural_area.plot.barh(left=settlement_area, color='orange')
wooded_area.plot.barh(left=settlement_area + agricultural_area, color='green')
unproductive_area.plot.barh(left=settlement_area + agricultural_area + wooded_area, color='red')

# annotate chart

plt.title('10 populated communes by type of area')
plt.xlabel('area type in %')
plt.legend(bbox_to_anchor=(1.05,1))
plt.show()
```

3. Histogram of young population

```python
# define list of bins for histogram

bins_list = [0, 5, 10, 15, 20, 25, 30, 35]

# plot histogram using specified bins list

hist = df['0-19 years'].plot.hist(bins=bins_list, figsize=(5,4), edgecolor='black')

# annotate chart

plt.title('Proportion of young residents by number of communes')
plt.xlabel('Percentage of residents')
plt.show()
```

4. Investigate distribution of 0-19 year and language across the regions

```python
#import seaborn for this section

import seaborn as sns

# create boxplot from the specified axis values within dataframe

sns.boxplot(x='Language', y='0-19 years', palette='pastel', data=df)

# annotate chart

plt.title('Distribution 0-19 years across the four spoken languages')
plt.show()
```

5. Investigate distribution of language across all age ranges

```python
# create a 3 by 1 chart with shared y axis and label

fig, ax = plt.subplots(1, 3, sharey=True, figsize=(20,7))
fig.suptitle('Distribution of ages across languages spoken')

# build individual charts using different age ranges and title

sns.boxplot(ax=ax[0], data=df, x='Language', y='0-19 years')
sns.stripplot(ax=ax[0], data=df, x='Language', y='0-19 years', palette='pastel', alpha=0.5)
ax[0].set_title('0-19 years')

sns.boxplot(ax=ax[1], data=df, x='Language', y='20-64 years')
sns.stripplot(ax=ax[1], data=df, x='Language', y='20-64 years', palette='pastel', alpha=0.5)
ax[1].set_title('20-64 years')

sns.boxplot(ax=ax[2], data=df, x='Language', y='65 years or over')
sns.stripplot(ax=ax[2], data=df, x='Language', y='65 years or over', palette='pastel', alpha=0.5)
ax[2].set_title('65 years or over')

plt.show()
```

6. Investigate relationship between land type of communes

```python
# extract subset of data required 

land_type = df[['Settlement area', 'Agricultural area', 'Wooded area', 'Unproductive area', 'Elevation']]

# plot pair using land types on x axis and adding elevation differences using hue 

pairplot = sns.pairplot(land_type, x_vars=['Settlement area', 'Wooded area', 'Unproductive area'], y_vars=['Agricultural area'], hue='Elevation')

# annote chart and show  

pairplot.fig.suptitle('Distribution profiles of land within Communes', y=1.1)
plt.show()
```

relationships above are not linear, simply showing the straight line as a representation of % of 100

7. Draw map of Switzerland from coordinates

```python
# extract required subset of data

coordinates = df[['East coordinate', 'North coordinate', 'Language', 'Elevation']]

# plot graph using elevation as the hue 

sp = sns.scatterplot(data=coordinates, x='East coordinate', y='North coordinate', hue='Elevation')

# annotate chart & show

sp.set_title('Commune map of Switzerland')
sp.legend(loc='upper left')
plt.show()
```

```python
# plot new graph using different hue

sp1 = sns.scatterplot(data=coordinates, x='East coordinate', y='North coordinate', hue='Language')

# annotate and show

sp1.set_title('Commune map of Switzerland with language')
sp1.legend(loc='upper left')
plt.show()
```

```python
# create two charts together using a shared y axis and title

fig, ax = plt.subplots(1, 2, sharey=True, figsize=(15,5))
fig.suptitle('Elevation vs language profiles of Switzerland')

# plot both graphs and label

ax1 = sns.scatterplot(ax=ax[0], data=coordinates, x='East coordinate', y='North coordinate', hue='Elevation')
ax1.set_title('Commune map with elevation')
ax1.legend(loc='upper left')

ax2 = sns.scatterplot(ax=ax[1], data=coordinates, x='East coordinate', y='North coordinate', hue='Language')
ax2.set_title('Commune map with language')
ax2.legend(loc='upper left')

plt.show()
```
