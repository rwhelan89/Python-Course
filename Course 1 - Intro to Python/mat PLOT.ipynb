```python
#default number of points is 50 if not specified
X = np.linspace(-np.pi, np.pi)
Y = np.sin(X) + np.cos(X)
```

```python
plt.plot(X, Y)
plt.show()
```

```python
plt.plot(X, Y, color="red", linewidth=2.0)
plt.show()
```

```python
plt.plot(X, Y, color="red", linewidth=2.0)
plt.xlim(-4, 4)
plt.show()
```

```python
#specify x ticks
plt.plot(X, Y, color="red", linewidth=2.0)
plt.xlim(-4, 4)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.show()
```

```python
#$ to format the mathematical symbols
plt.plot(X, Y, color="red", linewidth=2.0)
plt.xlim(-4, 4)
plt.xticks(
    [-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
    ["$-\pi$","$-\pi/2$", "$0$", "$\pi$", "$\pi/2$"],
)
plt.show()
```

```python
plt.plot(X, Y, color="red", linewidth=2.0)
plt.xlim(-4, 4)
plt.xticks(
    [-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
    ["$-\pi$","$-\pi/2$", "$0$", "$\pi$", "$\pi/2$"],
)
plt.title("The Function $f(x)=\sin(x) + \cos(x)$")
plt.xlabel("x")
plt.ylabel("y=f(x)")
plt.legend()
plt.show()
```

```python
X = np.linspace(-15,15)
f1 = np.sin(X)/X
f2 = 2*np.sin(X)/X
f3 = 3*np.sin(X)/X
plt.plot(X, f1, color="blue", linestyle='--', linewidth=2.0, label='$\sin(x)/x$')
plt.plot(X, f2, color="green", linestyle='-.')
plt.plot(X, f3, color="red", marker='o')
plt.ylim(-1, 3.5)
plt.yticks([-1,0,1,2,3])
plt.legend()
plt.show()
```
