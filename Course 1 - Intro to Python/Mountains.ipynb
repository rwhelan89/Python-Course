df.loc is a boolean function to bring up a [x.y].... adding a ~ to the first variable brings up the results that do NOT meet critera

```python
df.loc[high, "First ascent"]
```

```python
plt.hist(
    [df.loc[high, "First ascent"], df.loc[~high, "First ascent"]],
    stacked=True,
    edgecolor="black",
    bins=20,
)

plt.legend(["over 8000m", "under 8000m"], loc="upper right")
plt.xlabel("Year")
plt.ylabel("Number of first ascents")
plt.title("Year of first ascent")
plt.show()
```

clear from above that 1950-1970 were the super expedition age with all the high ascents
