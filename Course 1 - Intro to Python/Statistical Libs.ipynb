```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
sample_means = []
experiments = 100000

for i in range(experiments): #drawing 100000 samples each with 20 observations
    
    #drawing 20 obs from uniform(0,1) distro
    sample= np.random.uniform(0,1,size = 20)
    #taking mean and collecting them
    sample_means.append(np.mean(sample))
    
plt.hist(sample_means,bins=np.linspace(0,1,101));
plt.xlabel('sample mean')
plt.ylabel('frequency')
plt.show()
```

```python
#simpler usings stats

sample = stats.uniform.rvs(0,1,size = (20,100000))   #rvs is random variables
sample_means = np.mean(sample,0)         
                           
plt.hist(sample_means,bins=np.linspace(0,1,101));
plt.xlabel('sample mean')
plt.ylabel('frequency')
plt.show()                           
```

```python
#rolling dice in python

uniform_data = stats.uniform.rvs(1,6, size=10000).astype(int)

uniform_data
```

```python
plt.hist(uniform_data, bins=6)
plt.xlabel('random values')
plt.ylabel('frequency')
plt.show()
```

```python
#study average of every 5 rolls 
indexes = np.arange(0, 10000, 5)
print(indexes)
```

```python
print(len(indexes))
```

```python
# np.add.reduceat allows us to take running sums of successive values
sums = np.add.reduceat(uniform_data, indexes)
print(sums)
```

```python
# obtain avgs by dividing by 5
np.mean(sums / 5)
```

```python
plt.hist(sums / 5, bins=12)
plt.xlabel('averages')
plt.ylabel('frequency')
plt.show()
```
